```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 10)
font = {'family' : 'Arial',
        'size'   : 15}

plt.rc('font', **font)
plt.rcParams.update({'font.size': 25})
#plt.rcParams['axes.facecolor'] = '#D3D3D3'
```

```python
male_biker_counts = np.array([1405,3775,4886,4578,2113,544,67,7,12,4475]).astype(float)
female_biker_counts = np.array([348,941,1036,940,418,87,8,1,3,1265]).astype(float)
age_groups = ['Under 25','25-34','35-44','45-54','55-64','65-74','75-84','85-94','>=95','Unknown']
male_age_groups = pd.DataFrame({'group': age_groups, 'gender': ['Male']*10, 'counts': male_biker_counts})
female_age_groups = pd.DataFrame({'group': age_groups, 'gender': ['Female']*10, 'counts': female_biker_counts})
gender_age_map = male_age_groups.append(female_age_groups)
gender_age_map
```

```python
gender_age_map.columns = ['counts','Gender','group']
gender_age_map
```

```python
sns.set(rc={'figure.figsize':(15,10)},font_scale=2,style='white')
ax = sns.barplot(x="group", y="counts", hue="Gender", data=gender_age_map, palette = "Greys")
#ax.set_title('Age-Gender variation among Strava riders (2016)')
ax.set_ylabel('Number of Strava riders')
ax.set_xlabel("Age Groups")
plt.show()
```
